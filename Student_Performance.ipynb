Import Libraries

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

Load the Data

```python
df=pd.read_csv('students_performance.csv')
```

Show the Data

```python
df.head()
```

Check Basic info

```python
df.info()
```

Check Null Values

```python
df.isnull().sum()
```

**Handle Missing Values**

Fill Missing Value of Test Preparation with Mode(categorical nulls)

```python
df['Test Preparation'].fillna(df['Test Preparation'].mode()[0], inplace=True)
```

Fill Missing Value of Math Score with Mean(numeric nulls)

```python
df['Math Score'].fillna(df['Math Score'].mean(), inplace=True)
```

Fill Missing Value of Reading Score with Mean(numeric nulls)

```python
df['Reading Score'].fillna(df['Reading Score'].mean(), inplace=True)
```

Fill Missing Value of Writing Score with Mean(numeric nulls)

```python
df['Writing Score'].fillna(df['Writing Score'].mean(), inplace=True)
```

**Basic Analysis**

Compare average scores by gender

```python
print(df.groupby('Gender')[['Math Score', 'Reading Score', 'Writing Score']].mean())
```

Compare Average scores by test preparation

```python
print(df.groupby('Test Preparation')[['Math Score', 'Reading Score', 'Writing Score']].mean())
```

**Step 4: Visualization**

Count Plot: Students by City

```python
sns.countplot(data=df, x='City',palette='colorblind')
plt.title('Number of Students by City')
plt.xticks(rotation=45)
plt.show()
```

Count Plot: Students by Parental Education

```python
sns.countplot(data=df, x='Parental Education',palette='coolwarm')
plt.title('Parental Education Levels')
plt.xticks(rotation=30)
plt.show()
```

Histogram: Distribution of Math Scores

```python
plt.hist(df['Math Score'].dropna(), bins=10, color='pink', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

Violin Plot: Reading Score by School Type

```python
sns.violinplot(data=df, x='School Type', y='Reading Score', palette='Set2')
plt.title('Reading Score by School Type')
plt.show()
```

Average math score by gender using Bar Plot

```python
sns.barplot(data=df, x='Gender', y='Math Score',color='red')
plt.title("Average Math Score by Gender")
plt.show()
```

Score distribution by school type using Box plot

```python
sns.boxplot(data=df, x='School Type', y='Reading Score',color='green')
plt.title("Reading Score by School Type")
plt.show()
```

Correlation heatmap

```python
sns.heatmap(df[['Math Score', 'Reading Score', 'Writing Score']].corr(), annot=True, cmap='Blues')
plt.title("Score Correlation Heatmap")
plt.show()
```
