```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

Load Data

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill' , y='tip' , data=tips)
plt.title('Scatter Plot of Bill vs Tip')
plt.show()
```

```python
sns.scatterplot(x='total_bill' , y='tip' , data=tips , color='green','blue')
plt.title('Scatter Plot of Bill vs Tip')
plt.show()
#Need to Do
```

```python
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title('Histogram of Total Bills')
plt.show()
```

```python
sns.histplot(tips['tip'], bins=20, kde=True)
plt.title('Histogram of tip')
plt.show()
```

```python
sns.boxplot(x='day',y='tip', data=tips)
plt.title('Box Plot of Tpis by Day')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
sns.pairplot(tips)
plt.show
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips , color='red')
plt.title('Customized Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='day',y='total_bill',hue='time',data=tips,split=True)

plt.title("Violin Plot of Total Bills by Day and Time")
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='tip',y='total_bill',hue='time',data=tips,split=True)

plt.title("Violin Plot of Total Bills by Tip")
plt.xlabel('Tips')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='day', y='tip',hue='sex', data=tips , palette='colorblind' )
plt.title('Average Tip Amount by Day and Sex')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='day', hue='smoker',data=tips,palette='coolwarm')
plt.title('Count of diners by Day and Smoking Preference')
plt.xlabel('Day of the Week')
plt.ylabel('Count of Diners')
plt.legend(title='Smoker')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='day', hue='time',data=tips , palette='deep')
plt.title('Count of diners by Day and Time')
plt.xlabel('Days of the week')
plt.ylabel('Count of Diners')
plt.legend(title='Time')
plt.show()
```
